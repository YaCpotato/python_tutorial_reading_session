# 3. 数学の基礎
## 3.1.1 数式と記号
### 集合
ある要素$x$が、集合$S$に属していることは、次のような数式で表現で表現する

$$
x \in S
$$

集合の中身は波括弧{}を使って具体的に書き下すこともできる
$$
A = \{1,2,3,4 \} \\
B = \{2,4,6,8 \}
$$


```
a = set([1, 2, 3, 4])
b = set([2, 4, 6, 8])
print(f'A = {a}')
print(f'B = {b}')
```

例えば、$8 \in B$ですが、$8 \notin A$と書くと、集合に入っていないことも表現できる。  
2つの集合の共通部分には、次のような記号を使う。  
**積集合**
$$
A \cap B = \{2,4 \}
$$
**和集合**

$$
A \cup B = \{1,2,3,4,6,8 \}
$$

### 指数関数
$$
f(x) = 2^x
$$

```
import matplotlib.pyplot as plt
import seaborn as sns

import math
import numpy as np

plt.style.use('dark_background') # plt.style.availableで使えるスタイル
x = np.arange(-2, 5, 0.1)
y = math.e**x
#y = [math.e**i for i in x] 生Python

plt.plot(x, y)
plt.ylim(0, 140)
plt.xlabel('x', fontsize = 18)
plt.ylabel('f(x)', fontsize = 18)
```

### シグモイド関数

$$
f(x) = \frac{1}{1 + e^{-x}}
$$

```
x = np.arange(-8, 8, 0.1)
y = 1 / (1 + math.e ** (-1 * x))

plt.plot(x, y)
plt.ylim(0.0,1.0)
plt.xlabel('x', fontsize = 18)
plt.ylabel('f(x)', fontsize = 18)
```

### 対数関数
#### 自然対数
$$
f(x) = \log_{2}x
$$

2を底、$x$を真数という
底が2の対数のことを**自然対数**という

---

#### 常用対数
$$
f(x) = \log_{10}x
$$

底が10の対数のことを**常用対数**という

```
x = np.arange(0,8,0.1)
y = np.log2(x)

plt.plot(x, y)
plt.ylim(-4,3)
plt.xlabel('x', fontsize = 18)
plt.ylabel('f(x)', fontsize = 18)
```

### 双曲線関数

$$

\sinh x = \frac{e^x - e^{-x}}{2} \\
\cosh x = \frac{e^x + e^{-x}}{2} \\
\tanh x = \frac{\sinh x}{\cosh x} \\

$$


```
x = np.linspace(-np.pi, np.pi, 201)
tan_x = np.linspace(0, np.pi, 201)
sinh = np.sinh(x)
cosh = np.cosh(x)
tanh = np.tanh(x)

f = plt.figure(figsize=(10,3))
sinh_plot = f.add_subplot(131)
cosh_plot = f.add_subplot(132)
tanh_plot = f.add_subplot(133)

sinh_plot.plot(x, sinh)
cosh_plot.plot(x, cosh)
tanh_plot.plot(tan_x,tanh)
```
